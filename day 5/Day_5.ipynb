Write a Python Program to check if a number is a** Perfect Number**.*(Example:6=1+2+3)*

```python
num=int(input("Enter a number: "))
sum=0

for i in range(1,num-1+1,1):
  if num%i==0:
    sum+=1

if sum==num:
  print("Perfect Number")
else:
  print("Not a Perfect Number")
```

Write a Python Program to check if a number is a **Spy number\**.(Sum of digits = Product of digits)

```python
num = 1234
sum_of_digits=0
prod_of_digits=1
while num!=0:
  e=num%10
  num=num//10
  sum_of_digits=sum_of_digits+e
  prod_of_digits=prod_of_digits*e

if sum_of_digits==prod_of_digits:
  print("Spy Number")
else:
  print("Not a Spy Number")
```

Write a Python Program to check if a number is a **Buzz number**.(Divisible by 7 or ends with 7)



```python
num=1234567
last_digit=num%10

if num%7==0 or last_digit==7:
  print("Buzz Number ")
else:
  print("Not a Buzz Number")
```

Write a Python Program to check if a number is a **Strong Number**.(Example: 145 = 1! + 4! + 5!)

```python
def fact(num):
    fact=1
    for i in range(1,num+1,1):
      fact=fact*i
    return fact
num = 145
save=num
sum=0
while num!=0:
  e=num%10
  num=num//10
  a=fact(e)
  sum+=a
if save==sum:
  print("Strong Number")
else:
  print("Not a strong Number")
```
