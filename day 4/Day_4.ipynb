
write a Python program to generate the Fibonacci Series up to n terms.

```python
n=int(input("Enter nunber of terms: "))
a=0
b=1
for i in range(1,n+1,1):
  print(a,end=" ")
  sum=a+b
  a=b
  b=sum
```

Write a Python program to check whether a number is palindrome or not.


```python
n=1221
save=n
rev=0
while n!=0:
  e=n%10
  n=n//10
  rev=rev*10
  rev=rev+e
print(rev)

if rev==save:
  print("Palindrome")
else:
  print("Not a Palindrome")
```

write a Python program to print prime number from 1 to 100.

```python
n=100
for i in range(1,n+1,1):
  count=0
  for j in range(1,i+1,1):
    if i%j==0:
      count+=1
  if count==2:
    print(j,end=" ")
```

write a Python program for below patterns.

```python
n=5
for i in range(1,n+1,1):
  for j in range(1,i+1,1):
    print("*",end=" ")
  print()
```
