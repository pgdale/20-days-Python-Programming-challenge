Write a program to take N inputs from the user and store them in a list.

```python
n=int(input("SIZE: "))
store=[]
for i in range(1,n+1,1):
  choose=int(input("Enter: "))
  store.append(choose)
print(store)
```

Write a program to find the length of an list without using len().

```python
a=[1,2,3,4,5,6,7,8]
count=0
for i in a:
  count+=1
print(f"Length of list: {count}")
```

Write a program to find all the sum of the elements in a list.

```python
ele=[40,20,10,6,3,11]
sum=0
for i in ele:
  sum+=1
print(f"Sum of all elements in list: {sum}")
```

Write a program to Print only even numbers from an list.

```python
a=[1,45,67,89,12,31,44]
even=[]
for i in a:
 if i%2==0:
   even.append(i)
print(even)
```
